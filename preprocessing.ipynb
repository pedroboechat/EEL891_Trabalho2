```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv("data/conjunto_de_treinamento.csv")
df_test = pd.read_csv("data/conjunto_de_teste.csv")
print(list(df_train.columns))
df_train.head()
```

```python
tipos = list(set(df_train["tipo"].unique()).union(set(df_test["tipo"].unique())))
bairros = list(set(df_train["bairro"].unique()).union(set(df_test["bairro"].unique())))
tipos_bairros = tipos + bairros
```

```python
c = "tipo_vendedor"
print("Train:")
print(df_train[c].value_counts())
print("-"*20,"\nTest:")
df_test[c].value_counts()
```

```python
for column in df_train.columns:
    if column == "preco": continue
    print(column)
    print("NaN:", (np.nan in list(df_train[column].unique()) or np.nan in list(df_test[column].unique())))
    try:
        print("Same values:", (sorted(list(df_train[column].unique()))) == sorted(list(df_test[column].unique())))
        print("Diff:", [i for i in list(df_test[column].unique()) if (i not in (list(df_train[column].unique())) and column != "Id")])
    except: pass
    print(df_train[column].value_counts())
    print("\n\n\n\n")
```

```python
def treat_df(df, filename):
    selected_columns = tipos_bairros + ["churrasqueira", "estacionamento", "piscina", "playground", "quadra", "s_festas", "s_jogos", "s_ginastica", "sauna", "vista_mar", "quartos", "suites", "vagas", "area_util", "area_extra"]
    try:
        df = df[["tipo", "bairro", "churrasqueira", "estacionamento", "piscina", "playground", "quadra", "s_festas", "s_jogos", "s_ginastica", "sauna", "vista_mar", "quartos", "suites", "vagas", "area_util", "area_extra", "preco"]]
    except:
        df = df[["tipo", "bairro", "churrasqueira", "estacionamento", "piscina", "playground", "quadra", "s_festas", "s_jogos", "s_ginastica", "sauna", "vista_mar", "quartos", "suites", "vagas", "area_util", "area_extra"]]
    for tipo in tipos:
        df[tipo] = [1 if i == tipo else 0 for i in list(df["tipo"])]
    for bairro in bairros:
        df[bairro] = [1 if i == bairro else 0 for i in list(df["bairro"])]
    # df["tipo_vendedor"] = df["tipo_vendedor"].replace({"Imobiliaria": 1, "Pessoa Fisica": 0})
    df["area_extra"] = df["area_extra"].apply(lambda value: 1 if value != 0 else 0)
    df["vagas"] = df["vagas"].apply(lambda value: 5 if value > 4 else value)
    df["suites"] = df["suites"].apply(lambda value: 5 if value > 4 else value)
    df["quartos"] = df["quartos"].apply(lambda value: 5 if value > 4 else value)
    try:
        df = df[selected_columns + ["preco"]]
    except:
        df = df[selected_columns]
    df.to_csv(f"data/{filename}.csv", index=False)
    return df
```

```python
treat_df(df_train, "train")
```

```python
treat_df(df_test, "test")
```
